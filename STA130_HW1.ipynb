````python
# In this case, using df.dropna() is preferred because it allows you to remove rows with missing values without 
# losing any columns that might still be useful. If you used del df['col'], you would be removing entire columns, 
# which may not be desirable if those columns contain important data for your analysis.

# 2. Provide an example of "the opposite use case" in which using del df['col'] might be preferred over using 
#    df.dropna()

# ANSWER:
# CASE EXAMPLE:

# Dataset with several columns, and some columns are irrelevant or not useful for your analysis. 

import pandas as pd

# Example DataFrame
data = {'Name': ['Alice', 'Bob', 'Carol', 'Dave'],
        'Age': [30, 25, 35, 40],
        'Salary': [50000, 60000, 70000, 80000],
        'Extra_Info': ['X', 'Y', 'Z', 'W']}
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Remove the 'Extra_Info' column
del df['Extra_Info']

# Display the DataFrame after removing the column
print("\nDataFrame after removing 'Extra_Info':")
print(df)
````
